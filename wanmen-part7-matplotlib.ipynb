# Histogram

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

np.random.seed(2019)
mu,sigma = 100,15
x = mu+sigma*np.random.randn(10000)
n,bins,patches = plt.hist(x,50,normed = 1,facecolor='b',alpha=0.5)
print('n=',n)
print('bins=',bins)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,.025,r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

# patches

```python
x = np.random.normal(size=100)
print(x)
n,bins,patches = plt.hist(x)
plt.setp(patches[0],'facecolor','r',alpha=0.5)
plt.show()
```

```python
np.random.seed(19680801)

dt = 0.001
t = np.arange(0.0,10.0,dt)
r = np.exp(-t[:1000] / 0.05)
x = np.random.randn(len(t))
s = np.convolve(x,r)[:len(x)] * dt


plt.plot(t,s)
plt.axis([0,1,1.1 * np.min(s),2 * np.max(s)])
plt.xlabel('time(s)')
plt.ylabel('current(nA)')
plt.title('Gausiau colored noise')


a = plt.axes([.65,.6,.2,.2],facecolor='k')
n,bins,patches = plt.hist(s,40,normed=1)
plt.title('Probability')
plt.xticks([])
plt.yticks([])


a = plt.axes([0.2,0.6,.2,.2],facecolor='k')
plt.plot(t[:len(r)],r)
plt.title('Impuse response')
plt.xlim(0,0.02)
plt.xticks([])
plt.yticks([])

plt.show()
```

# 3D图

```python


import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

```

```python
X = np.arange(-5,5,0.25)
Y = np.arange(-5,5,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X ** 2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.viridis)
plt.show()

```
